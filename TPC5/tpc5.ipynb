```python

import pandas as pd
import csv
def TabelaExcel():
    ficheiro=pd.read_csv('myheart.csv')
    print(ficheiro)

tabela=[]

def lerinformação():
    ficheiro = open('myheart.csv', mode='r')
    inftabela= csv.reader(ficheiro,delimiter=',')
    nlinha=0
    for linha in inftabela:
        if nlinha==0:
            nlinha+=1
        else:
            tabela.append((int(linha[0]),linha[1],int(linha[2]),int(linha[3]),int(linha[4]),int(linha[5])))

def distribuicaosexo():
    lerinformação()
    contadorM=0
    contadorF=0
    for t in tabela:
        idade,sexo,tensao,colestrol,batimento,temd=t
        if temd==1 and sexo== "M":
            contadorM+=1
        elif temd==1 and sexo=="F":
            contadorF+=1
    print("Contagem masculina:", contadorM)
    print("Contagem feminina:", contadorF)
  
    

def distribuiçãoidade():
    lerinformação()
    escalões=[[20,24],[25,29],[30,34],[35,39],[40,44],[45,49],[50,54],[55,59],[60,64],[65,69],[70,74],[75,79],[80,84],[85,89]]
    lista=[]
    for t in tabela:
        idade,sexo,tensao,colestrol,batimento,temd=t
        for escalao in escalões:
            if temd==1 and idade>=escalao[0] and idade <=escalao[1]:
                lista.append(escalao)
    for escalao in escalões:
        print("Escalão ",escalao," tem:",lista.count(escalao)," pessoas")

def distribuiçãocolestrol():
    lerinformação()
    lista=[]
    nivelcolest=[]
    i=0
    b=54
    while b>0:
        p=i+10
        n=[i,p]
        nivelcolest.append(n)
        i=i+10
        b=b-1  
    for t in tabela:
       idade,sexo,tensao,colestrol,batimento,temd=t
       for nivel in nivelcolest:
        if temd==1 and colestrol>= nivel[0] and colestrol <=nivel[1]:
            lista.append(nivel)
    for nivel in nivelcolest:
        print("Nivel",nivel,"tem:", lista.count(nivel),"pessoas")

def tabular():
    num=int(input("Escolha a distribuição que pretende tabular:\n 1)sexo \n 2)idade \n 3)colestrol"))
    if num==1:
        print("   M        F   ")
        print("-----------------")
        print("  428       40    ")          #ERRADOOOOO!
    elif num==2:
        print("Escalão      Casos de doença")
        print("-----------------------------")
        print("NÃO SEI, ESTÁ MALLLLLLL, ERRADOOO")


def menu():
    opção= int(input("introduza 1 nº entre 0 e 5"))
    if opção==0:
        TabelaExcel()
    elif opção==1:
        lerinformação()
    elif opção==2:
        distribuicaosexo()
    elif opção ==3:
        distribuiçãoidade()
    elif opção==4:
        distribuiçãocolestrol()
    elif opção==5:
        tabular()

menu()
```
